# Pie Chart

```python
import matplotlib.pyplot as plt

# Parameters for the pie chart
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [25, 35, 20, 20]  # Corresponding sizes of the pie slices
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']  # Colors for each slice
startangle = 140  # Starting angle of the pie chart
title = 'Sample Pie Chart'  # Title of the pie chart
caption = 'This is a sample caption for the pie chart'  # Caption for the pie chart

# Custom function to display the numbers alongside the percentages
def func(pct, allvals):
    absolute = int(pct/100.*sum(allvals))
    return f"{absolute} ({pct:.1f}%)"

# Create a figure and a set of subplots
fig, ax = plt.subplots(dpi=300)

# Plotting the pie chart
wedges, texts, autotexts = ax.pie(sizes, labels=labels,
                                  autopct=lambda pct: func(pct, sizes), startangle=startangle)

# Equal aspect ratio ensures that the pie is drawn as a circle
ax.axis('equal')

# Adding a title
ax.set_title(title)

# Adding a caption
fig.text(0.5, 0.02, caption, ha='center')
```

# Stacked Bar Chart

```python
import matplotlib.pyplot as plt

# Parameters for the stacked bar chart
categories = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [25, 35, 20, 20]  # Corresponding sizes for each category
title = 'Sample Horizontal Stacked Bar Chart'
caption = 'This is a sample caption for the horizontal stacked bar chart'

# Calculate total size for percentage calculation
total_size = sum(sizes)

# Create a figure and a set of subplots
fig, ax = plt.subplots(figsize=(10, 1), dpi=300)

# Plotting the single horizontal stacked bar chart
left = 0
bar_height = 0.2
for i, size in enumerate(sizes):
    ax.barh('Total', size, left=left, height=bar_height ,label=categories[i])
    # Adding the percentage text
    percentage = size / total_size * 100
    ax.text(left + size / 2, 0, f'{size} ({percentage:.1f}%)', va='center', ha='center', color='black')
    left += size

# Adding a title
ax.set_title(title)

# Adding labels
ax.set_xlabel('Values')

# Adding a caption
fig.text(0.5, -0.7, caption, ha='center')
```
